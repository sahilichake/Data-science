# One_Hot_Encoder

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe = ohe.fit_transform(cars_cat[["num-of-doors"]]).toarray()
ohe_col = pd.DataFrame(ohe)
cars_cat = pd.concat([cars_cat,ohe_col],axis=1)
cars_cat = cars_cat.drop(['num-of-doors'],axis=1)
```

```python
cars_cat
```

# Label_Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(cars["fuel-type"])
```

```python
for col in cars_cat:
    le=LabelEncoder()
    cars_cat[col]=le.fit_transform(cars_cat[col])
```
