```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tr = pd.read_csv("housing_prices_train.csv")
```

```python
tr["SalePrice"].describe()
sns.displot(tr["SalePrice"],color="r")
```

```python
tr["SalePrice"].describe()
sns.distplot(tr["SalePrice"],color="r")
print('Skew of Rawadata : %f'%tr['SalePrice'].skew())
```

# Log Transformation

```python
Singlelog_y = np.log1p(tr['SalePrice'])
```

```python
tr["SalePrice"].describe()
sns.distplot(Singlelog_y,color="r")
print('Skew after 1st log transform : %f'%Singlelog_y.skew())
```

```python
Doublelog_y = np.log1p(Singlelog_y)
sns.distplot(Doublelog_y,color="r")
print('Skew after 2st log transform : %f'%Doublelog_y.skew())
```

# Square Root Transform 

```python
square = np.sqrt(tr['SalePrice'])
tr["SalePrice"].describe()
sns.distplot(square,color="r")
print('Skew after Square Root Transform : %f'%square.skew())
```

# Cube Root Transform

```python
cube = np.cbrt(tr['SalePrice'])
tr["SalePrice"].describe()
sns.distplot(cube,color="r")
print('Skew after Cube Root Transform : %f'%cube.skew())
```
